Visualisation on the basis of two values
```
. Categorical data
. Numerical data

Basic Graph Types
```
-Line graph : show trends over time
-Bar graph : compare values b/w categories (discrete data)
-Pie chart : compare values b/w categories (percentage data)
-Histogram : show distribution of numerical data 
-Box plot : statistical distribution of numerical data 
-Heat map : show matrix of values using color
-Scatter plot : show relationship b/w two variables 
-Bubble chart : show relationship b/w three variable

````python
sns.countplot(df, x='class', hue="who")
plt.title('passengers count by age')
plt.xlabel("class")
plt.ylabel("count")
plt.show()
````

````python
sns.histplot(df, x='class')
plt.title('passengers count by age')
plt.xlabel("class")
plt.ylabel("count")
plt.show()
````

````python
sns.histplot(df, x='age', bins=10, kde= True)
plt.title('passengers count by age')
plt.xlabel("age")
plt.ylabel("count")
plt.show()
````
